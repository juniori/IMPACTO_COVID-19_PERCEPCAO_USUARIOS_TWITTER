Fonte: https://infoms.saude.gov.br/extensions/covid-19_html/covid-19_html.html (datasets baixados em 07/03/2022)


```python
import pandas as pd
base_path = '_Mestrado/SEAN/Dissertacao_Celso/resources/datasets/HIST_PAINEL_COVIDBR_06mar2022/'
dados_primeiro_semestre = base_path + 'HIST_PAINEL_COVIDBR_2021_Parte1_06mar2022.csv'  
dados_segundo_semestre = base_path + 'HIST_PAINEL_COVIDBR_2021_Parte2_06mar2022.csv'  


# Li os arquivos com os dados do 1º e 2º Semestre de 2021
df_ps = pd.read_csv(dados_primeiro_semestre,sep=';')
df_ss = pd.read_csv(dados_segundo_semestre,sep=';')

# Os dois arquivos foram concatenados (union all do SQL) e utilizamos os registros indicados pela região Brasil.
df = pd.concat([df_ps[df_ps.regiao == 'Brasil'],df_ss[df_ss.regiao == 'Brasil']])

# Crio uma com o mês extraído da data para em seguida agrupar por mês
df['mes'] = pd.DatetimeIndex(df['data']).month
```

```python

# Cada arquivo possui 17 colunas 
# regiao;estado;municipio;coduf;codmun;codRegiaoSaude;nomeRegiaoSaude;data;semanaEpi;populacaoTCU2019;casosAcumulado;casosNovos;obitosAcumulado;obitosNovos;Recuperadosnovos;emAcompanhamentoNovos;interior/metropolitana
df.shape
```

```python

# Filtrando por um mês específico para validar os dados
# 29 de abril: Brasil atinge 400 mil mortes por Covid-19, com 3.001 novos óbitos registrados neste dia, segundo o Ministério da Saúde.

df_mes = df[df.mes==4]
df_mes.head()
df_mes = df_mes.groupby(['mes']).agg({'casosNovos': ['sum'],'obitosNovos': ['sum']})

df_mes.head(12)


```
